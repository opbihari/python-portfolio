**4.Problem Statement:**
Write a lambda function to check if a given string is a palindrome. The lambda function should return True if the string is the same when read forwards and backwards, and False otherwise.

**Expected Output:**If the string is a palindrome, the output should be True, otherwise False.

**Example:**

Input: "racecar" Output: True
* "racecar" is a palindrome because if you read it from left to right or right to left, it’s the same.

Input: "hello" Output: False
* hello" → when read from left to right, it's "hello", but when read backward, it's "olleh". This is not a palindrome.

**Explanation:**A palindrome is a word or phrase that reads the same backward as forward, ignoring spaces, punctuation, and capitalization.


```python
input_string = input("Enter a string: ")
is_palindrome = lambda x: x == x[::-1]
print(is_palindrome(input_string))
```

Example:

**6.Problem Statement:**Write a lambda function that takes a number as input and checks if it is divisible by both 2 and 3. The function should return True if the number is divisible by both, otherwise return False.

**Expected Output:**
The output should be True if the number is divisible by both 2 and 3, otherwise False.

**Example:**
* Input: 6
* Output: True
* (6 is divisible by both 2 and 3)

* Input: 8
* Output: False
* (8 is divisible by 2 but not by 3)

```python
inp_number = int(input("Enter a number: "))
is_divsible = lambda x: True if x % 2 == 0 and x % 3 == 0 else False
print(is_divsible(inp_number))
```

**7.Problem Statement:**
Write a program that uses the map() function along with a lambda function to multiply each element in the given list by 2. The output should be a new list with each element multiplied by 2.

**Expected Output:**
The output should be a list where each number is multiplied by 2.

Example:

* Input: [1, 2, 3, 4, 5]
* Output: [2, 4, 6, 8, 10]

* Input: [10, 20, 30]
* Output: [20, 40, 60]

```python
multi_2 = lambda input_list: map(lambda x: x*2, input_list)
print(list(multi_2([1,2,3,4,5])))
```
