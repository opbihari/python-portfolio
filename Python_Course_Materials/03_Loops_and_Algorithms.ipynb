*Problem Statement:Skip Multiples of 3**

Write a while loop that counts from 1 to 10 but skips printing numbers that are multiples of 3. Use continue.

Expected Output: 1, 2, 4, 5, 7, 8, 10


```python
# Initialize a counter
count = 1

# Loop from 1 to 10 using a while loop
while count <= 10:
  # Check if the current number is a multiple of 3
  if count % 3 == 0:
    # If it's a multiple of 3, increment and skip to the next iteration
    count += 1
    continue
  # If it's not a multiple of 3, print the number
  print(count)
  # Increment the counter for the next iteration
  count += 1
```

You are given a list of numbers. Start by adding them one by one to a total sum. If, at any point, the sum exceeds 30, stop adding further numbers. However, do not include the number that caused the sum to exceed 30. Only print the sum of numbers that were added before it exceeded 30.

* Add each number from the list to a running total (starting from 0).
* As soon as adding a number causes the total to go over 30, stop and do not count that number.
* Print the sum right before it exceeds 30.

Input:
A list of numbers, for example: [5, 10, 15, 10, 20]

Expected Output:
The total sum just before it exceeds 30. If the sum goes over 30 with the addition of a number, that number should not be counted in the final sum.

Note:If the sum exceeds 30 after adding a number, we subtract that number (total -= num) to undo the addition.

```python
list_of_numbers = [5,10,15,20]
sum_of_numm = []
sum = 0
for c in list_of_numbers:
  if sum < 30:
    sum += c
    sum_of_numm.append(c)
  else:
   break

print(f"sum of number {sum_of_numm} is {sum}")


```
