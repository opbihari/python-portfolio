**Problem Statement1:**

Create a program that takes two numbers as input  and converts the input values to floating-point numbers using float() to handle both integers and decimals, stores them in two variables, adds them, and prints the result using a formatted string.

**Example Input:**

 * Enter the first number: 5.0  

 * Enter the second number: 10.0

  * **`Example Output:`** The sum of 5.0 and 10.0 is: 15.0


```python
#Solution 1

#first number
first_number  = float(input("Enter the first number: "))

#second n8umber
second_number = float(input("Enter the second number: "))

#sum of two number
sum_of_two_numbers = first_number + second_number

print(f"The sum of {first_number} and {second_number} is: {sum_of_two_numbers}")
```

# Problem Statement 2:

 Create a program that takes two numbers as input from the user and converts the input values to floating-point numbers using float() to handle both integers and decimals , stores them in two separate variables, multiply the first and second number, and prints the result using a formatted string.

**`Example Input:`**

 * Enter the first number: 4.0`

 * Enter the second number: 5.0

 **`Example Output:`** :The product of 4.0 and 5.0 is:  20.0


```python
#Solution 2

#first number
first_number  = float(input("Enter the first number: "))

#second number
second_number = float(input("Enter the second number: "))

#multiplying two number stored in two variable
product_of_two_numbers = (first_number * second_number)

#printing result in formatted string
print(f"The product of {first_number} and {second_number} is {product_of_two_numbers}")
```

# Problem Statement 4:

Create a Python program that takes two numbers as input from the user and converts the input values to floating-point numbers using float() to handle both integers and decimals, stores them in two variables, and uses a condition to check if both numbers are greater than zero (positive) using the and operator. Print the result using a formatted string to clearly display both numbers and whether they are positive or not.

**`Example Input:`**

 * Enter the first number: -2.0

  * Enter the second number: 5.0

**`Example Output:`** Are both -2.0 and 5.0 positive? False

```python
#solution 4

#first number
first_number  = float(input("Enter the first number: "))

#second number
second_number = float(input("Enter the second number: "))

#checking if both numbers are greater than zero
comparison = (first_number > 0 and second_number > 0)

#printing result using formatted string
print(f"Are both {first_number} and {second_number} positive? {comparison}")
```

# Problem Statement 5:  

Create a program that takes three numbers as input from the user  and converts the input values to floating-point numbers using float() to handle both integers and decimals, stores them in three separate variables, calculates their average by adding all the three numbers and then divide by 3, and prints the result using a formatted string.

**`Example Input:`**

 * Enter the first number: 10.0

 * Enter the second number: 20.0

 * Enter the third number: 30.0

**`Example Output:`** The average of 10.0, 20.0, and 30.0 is: 20.0

```python
#solution 5

#first number
first_number  = float(input("Enter the first number: "))

#second number
second_number = float(input("Enter the second number: "))

#third number
third_number = float(input("Enter the third number: "))

#calculating average by adding all three and dividing them by 3
average = (first_number + second_number + third_number) / 3

#printing result using formatted string
print(f"The average of {first_number}{second_number} and {third_number} is {average}")
```
