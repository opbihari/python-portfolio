**1. Calculate Total Bill with Tax and Tip**

>  **Problem:** Write a program that takes the bill amount, tax percentage, and tip percentage as inputs using input function and store in 3 different variables  and calculates the final bill amount.

**Example Output:**
 * Enter bill amount: 100  
 * Enter tax percentage: 10  
 * Enter tip percentage: 15  
 * Total bill: 125.0

```python
#SOLUTION

#bill
bill_amount = float(input("Enter bill amount: "))

#tax
tax_percentage = float(input("Enter tax percentage: "))

#tip
tip_percentage = float(input("Enter tip percentage: "))

tax_amount = (tax_percentage / 100) * bill_amount

tip_amount = (tip_percentage / 100) * bill_amount

total_bill = bill_amount + tax_amount + tip_amount

print(" your Total bill:", total_bill)
```

Write a program to calculate the ticket price for a movie in rupees based on the person’s age and the time of the show (morning or evening). Use the input() function to take both the age and showtime as inputs from the user.

The ticket pricing rules are as follows:

Ticket Prices by Age Group:

* Child (12 years old or younger): ₹100
* Adult (between 13 and 59 years old): ₹240
* Senior (60 years or older): ₹160

Discount Based on Showtime:

* Morning Show: If the showtime is "morning", apply a 20% discount to the ticket price.
* Evening Show: If the showtime is "evening", there is no discount applied.

**Example**

 > Example Input:

age = 30, showtime = 'morning'

 > **Expected Output:**

Ticket Price: ₹192.0

**Explanation:**
In this example, the person is 30 years old, which falls into the adult category, so the base ticket price is ₹240. Since the showtime is in the morning, a 20% discount is applied:

Discounted Price = ₹240 - (20% of ₹240) = ₹192

If the showtime was "evening" for the same age, the ticket price would remain ₹240 as no discount applies.

```python
#SOLUTION 7

customer_name = input("Enter your Name")   # costumer name

#costomer name

age = int(input("Enter ypur age "))

#timing
show_time = input("enter show time morning or evening  --").lower()

if age <= 12:
  price = 100

elif 13 <= age <= 59:
  price = 240

else:
  price = 160

if show_time == "morning":
  final_price = (price - (price / 100 * 20))
  print( customer_name,"YOUR TOTAL WILL BE :",final_price )

else:
  print(customer_name"YOUR TOTAL WILL BE :",price )
```

#**8. Restaurant Bill Splitter with Service Charge**

>You and your friends went to a restaurant and received the final bill. The restaurant adds a service charge on top of the total. You need to find out how much each friend should pay after including the service charge.


Write a Python program that takes the total bill amount, service charge percentage, and the number of friends as input, and calculates the amount each friend must pay.

**Input Format:**

1. Total bill amount (float)

2. Service charge percentage (float)

3. Number of friends (integer)

**Output Format:**

>Print “Each pays: amount” (rounded to 2 decimals)

**Example Input:**
1. Enter total bill: 1200
2. Enter service charge percentage: 5
3. Enter number of friends: 3

**Expected Output:**

>Each pays: 420.0


```python
#solution 8

bill = float(input("Enter total bill: ")) #initial bill

service_charge_prcentage = float(input("Enter service charge percentage: ")) #service charge

friends = int(input("Enter number of friends: ")) #number of friends

service_charge = (service_charge_prcentage / 100) * bill #service charge

total_bill = bill + service_charge

each_pays = total_bill / friends

print("Each pays:", round(each_pays, 2))
```

#**11. Gym Membership Discount Calculator**

A gym offers membership discounts based on age and joining time. Younger or senior members get special offers, and morning joiners get an extra discount.

Write a program to calculate the final membership fee after applying discounts.

**Rules:**

1. Base fee = ₹1500

2. If age < 21 or age ≥ 60 → 15% off

3. If joining in the morning → extra 5% off on the discounted price

**Input Format:**

1. Age (int)

2. Time of joining (string: morning/evening)

**Output Format:**

>Final fee rounded to 2 decimals.

**Example Input:**
1. Enter age: 19
2. Enter joining time (morning/evening): morning

**Expected Output:**

`Final fee: 1211.25`

```python
#solution 11

age = int(input("Enter age: ")) #age

time_of_joining = input("Enter joining time (morning/evening): ").lower() #time of joining


base_fee = 1500

if age < 21 or age >= 60:
    discount = base_fee*0.15

else:
    discount = 0

if time_of_joining == "morning":
        discunted_price = base_fee - discount
        final_fee = discunted_price - (discunted_price * 0.05)
        print("Final fee:", round(final_fee, 2))

else:
        final_fee = base_fee - discount
        print("Final fee:", round(final_fee, 2))
```
