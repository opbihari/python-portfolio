* Print the common ingredients so you can see which items are shared between the two recipes.

Key Points:

### **5. Problem: Find All Items with Prices Above ₹100**
Imagine you have a dictionary where each fruits in a store is listed as a key, and its price in rupees is the value. You want to find and display only the Fruits that cost more than ₹100. This can help you focus on the more expensive items and filter out the lower-priced ones.

In this problem, you need to:

* Initialize a new empty dictionary called above_hundred_rupees to store fruit priced above ₹100.
* Use a for loop to go through each item and its price in the original prices dictionary.
* Use an if condition to check if the price of each item is greater than ₹100.
* If the condition is met, add that item and its price to the above_hundred_rupees dictionary.
* Print the new dictionary that contains only the items priced above ₹100.

Key Points:
* A dictionary stores key-value pairs, where each key is associated with a specific value.
* Use a for loop with the .items() method to access both the keys and values.
* The if condition allows filtering of items based on their prices.

**Sample Output:**
Items above ₹100: {'apple': 120, 'cherry': 250, 'dates': 150}

```python
list_of_fruits = {
     100 :'mango',
     120 :'apple',
     250 :'cherry',
     150 : 'dates',
}
above_hundred_rupees = {}

for fruit_dam , neem in list_of_fruits.items():
  if fruit_dam > 100:
    above_hundred_rupees[neem] = fruit_dam

print(f"Items above ₹100 : {list(above_hundred_rupees)}")
```

### **10. Problem: Add Multiple Products to a Store Inventory**

Imagine you are managing a store’s inventory, which is represented by a set of products currently available. You want to add new products to this inventory as they arrive. You’ll keep adding products one by one until you’re done, and then display the updated inventory.

In this problem, you need to:

* Start with an existing set of products in the inventory.
* Prompt the user to enter the name of a new product to add to the inventory.
* If the user types the product name, add it to the inventory set.
* Repeat this process until the user types "done," indicating they’ve finished entering new products.
* Print the updated inventory set to show all the products now available.

Key Points:
* Sets automatically store unique items, so any product entered more than once will only appear once in the inventory.
* Using a while loop allows continuous input until the user decides to stop by typing "done."
* The break statement is used to exit the loop when the user indicates they’re finished.

**Sample Output:**

Enter a product to add to inventory (type 'done' to stop): grapes

Enter a product to add to inventory (type 'done' to stop): mango

Enter a product to add to inventory (type 'done' to stop): done

Updated Inventory: {'apple', 'banana', 'orange', 'grapes', 'mango'}

### **11.Problem: Manage Restaurant Reservations**
Imagine you are managing reservations for a restaurant, where each table has a specific reservation. You have a dictionary where each key is a table name (e.g., "Table1") and each value is the name of the guest who reserved it. You want to give the user two options: they can either add a new reservation or remove an existing reservation.

In this problem, you need to:

* Start with a dictionary that represents current reservations.
* Prompt the user to choose an action:

   "add" to add a reservation.

    "remove" to remove an existing reservation.

-- If the user chooses to add:

* Ask them to enter the table name and guest name.

-- Check if the table already has a reservation:
* If it does, inform the user that the table is already reserved (e.g., "Table1 already has a reservation for Alice").
* If it does not, update the dictionary by adding the guest name for the specified table and confirm the addition.

--If the user chooses to remove:
* Ask them to enter the table name of the reservation they want to remove.
Check if the table has an existing reservation:
* If it does, delete it and confirm the removal (e.g., "Reservation for Table1 removed").
* If it does not, inform the user that no reservation was found for the specified table.
* Finally, display the current list of reservations to show the updated reservations.

Key Points:
* Dictionaries are ideal for managing reservations as they allow you to associate each table with a specific guest’s name.
* if and elif statements enable handling different actions based on the user’s input.
* The in keyword checks if a specific table exists in the dictionary, while the del statement removes a key-value pair if the table is found.

**Sample Output:**

Adding a reservation (table already reserved):

Enter 'add' to add a reservation or 'remove' to remove a reservation: add

Enter table name: Table1

Table1 already has a reservation for Alice.

Current Reservations: {'Table1': 'Alice', 'Table2': 'Bob'}

Removing a reservation (table not found):

Enter 'add' to add a reservation or 'remove' to remove a reservation: remove

Enter table name to remove reservation: Table3

No reservation found for Table3.

Current Reservations: {'Table1': 'Alice', 'Table2': 'Bob'}

```python
reservation_dict = {
    'Table1':'Alice',
    'Table2':'Bob',
}
user_input = input("Enter 'add' to add a reservation or 'remove' to remove a reservation: ").lower()

if user_input == 'add':
  table_name = input("Enter table name: ")
  if table_name in reservation_dict:
    print(f"{table_name} already has a reservation for {reservation_dict[table_name]}.")
  else:
    guest_name = input("Enter guest name: ").lower()
    reservation_dict[table_name] = guest_name
    print(f"{table_name} has been reserved for {guest_name}")
elif user_input == 'remove':
  table_name = input("Enter table name to remove reservation: ")
  if table_name in reservation_dict:
    del reservation_dict[table_name]
    print(f"Reservation for {table_name} removed.")

  else:
    print(f"No reservation found for {table_name}.")


print(f"Current Reservations : {reservation_dict}")
```

### **12. Problem: Update Product Prices**
Imagine you are managing a store’s inventory, which includes a dictionary of products and their prices. Each key in the dictionary is the name of a product (like "apple" or "banana"), and each value is its price.
You want to give the user an option to update the price of a specific product. If the product is in the inventory, you’ll allow them to enter a new price to replace the old one. If the product isn’t in the inventory, you’ll let the user know that the product wasn’t found.

In this problem, you need to:

* Start with a dictionary that contains the names and prices of products.
* Prompt the user to enter the name of the product they want to update.
* Use an if statement to check if the product is in the dictionary.
1. If it is, ask the user for the new price, update the dictionary with this new price, and confirm the update.
2. If it isn’t, inform the user that the product is not in the inventory.
* Display the updated dictionary to show the current prices of all products.

Key Points:

* Dictionaries are ideal for pairing items with specific information, like products with prices.
* Using if and else statements allows you to handle cases where the product is in the dictionary or not.
* Directly updating the value for a key in a dictionary will replace the existing value with the new one.

**Sample Output:**

Enter the product to update price: apple

Enter the new price for apple: 60

Price for apple updated to 60.

Updated Prices: {'apple': 60, 'banana': 30, 'cherry': 100}

**Or, if the product is not in the inventory:**

Enter the product to update price: grapes

grapes not found in inventory.

Updated Prices: {'apple': 50, 'banana': 30, 'cherry': 100}


```python
store_inventory = {
    'apple' : 50,
    'banana' : 30,
    'cherry' : 20,}

product = input("Enter the product to update price: ")

if product in store_inventory:
  new_price = int(input("Enter the new price for apple: "))
  store_inventory[product] = new_price
  print(f"Price for {product} updated to {new_price}.")

else:
  print(f"{product} not found in inventory")

print(f"Updated price : {store_inventory}")
```

### **15.Problem: Manage Friends’ Birthdays**
Imagine you want to keep track of your friends’ birthdays, so you create a dictionary where each friend’s name is the key, and their birthday date is the value.

In this task, you will:

* Start with a dictionary that already contains some friends’ birthdays.
* Prompt the user to enter the name of a friend.
* Check if the friend’s name is already in the dictionary:
* If the friend’s name exists in the dictionary, display their current birthday and ask if the user wants to update it:
* If the user chooses to update, prompt them for the new date and update the dictionary with this new date.
* If the user does not want to update, keep the existing date.
* If the friend’s name is not in the dictionary, ask the user to enter the birthday date and add it as a new entry.
* Display the updated dictionary showing all current friends and their birthdays, including any additions or updates.

Key Points:
* Dictionaries are useful here because they allow quick access to each friend’s name and birthday.
* User interaction provides the option to update an existing birthday or retain it.
* Assigning a value to a dictionary key updates it if the key exists or adds it if the key does not exist.

**Sample Output:**
If you update an existing friend’s birthday:

Enter the friend's name: Alice

Alice is already added with a birthday on 12th Jan.

Would you like to update the birthday? (yes/no): yes

Enter the new birthday date: 10th Jan

```python
friend_list = {
    'Alice' : '12th Jan',
    'Bob' : '3rd Feb',
    'Charlie' : '15th Mar',
}
friend_name = input("Enter the friend's name: ")

if friend_name in friend_list:
  print(f"{friend_name} is already added with a birthday on {friend_list[friend_name]}.")
  update = input("Would you like to update the birthday? (yes/no): ").lower()

  if update == 'yes':
    new_birthday = input("Enter the new birthday date: ")
    friend_list[friend_name] = new_birthday
    print(f"{friend_name}'s birthday has been updated to {new_birthday}.")
  else:
    print(f"{friend_name} has not been changed")

else:
  bithday = input("Enter the birthday date: ")
  friend_list[friend_name] = bithday
  print(f"Added {friend_name}'s birthday on {bithday}.")

print(f"all friends birth day are {friend_list}")
```
