**Problem Statement1:**

Create a program that takes a person’s age as input and prints whether they get a senior citizen discount (age >= 60) or not using if and else statement only.

 * Example Input and Output:

 * Input:
  :-Enter your age: 65

**`Output:`** You are eligible for a senior citizen discount.

```python
#solution 1

age = int(input("Enter your age: "))

if age >= 60:
    print("You are eligible for a senior citizen discount.")

else : print("You are not eligible for a senior citizen discount.")
```

# **Problem Statement2:**

Create a Python program that takes three numbers as input from the user, stores them in three separate variables and converts the input values to floating-point numbers using float() to handle both integers and decimals , and checks which number is the largest among the three. Use comparison operators (>=) along with the and operator to ensure the correct logic for identifying the largest number. If two or more numbers are equal and also the largest, the program should still print that number. Display the result using a formatted string.

**Example Input and Output:**

 * Input:

   :- Enter the first number: 10

   :- Enter the second number: 20

   :- Enter the third number: 15

 **`Output:`** The largest number is: 20.0

```python
#solution 2

first_number = float(input("Enter the first number: "))

second_number = float(input("Enter the second number: "))

third_number = float(input("Enter the third number: "))

#checking first number
if first_number >= second_number and first_number >= third_number:
   print(f"largest_number = {first_number}")

#checking second number
elif second_number >= first_number and second_number >= third_number:
      print(f"largest_number = {second_number}")


else: print(f"largest_number = {third_number}")

```

#**Problem Statement3:**

Create a Python program that takes the total bill amount as input from the user. If the total bill is greater than ₹1000, the program should calculate a 10% discount by multiplying the total bill by 0.1 (10%) and subtracting the discount from the total bill. If the bill is ₹1000 or less, no discount is applied. The program should print the final amount using a formatted string.

`Example Input and Output:`

  * Input:

   :- Enter the total bill amount: ₹1200

 `Output:`

Discount applied. Final amount: ₹1080.0

```python
#solution 3

total_bill = float(input("Enter the total bill amount: ₹"))

discount = total_bill*10/100

discounted_amount = total_bill - discount

if total_bill > 1000:
  print(f"Discount applied.\
   Final amount: ₹{discounted_amount}")

else:
   print("No discount applied")
```

Create a Python program that stores predefined values for username and password. The program should prompt the user to enter both a username and a password, and check if both inputs match the stored values using the and operator. If both the username and password match exactly, the program should print "Login successful." Otherwise, it should print "Invalid credentials." You must use if and else statements to handle both scenarios. Store the predefined (stored) username and password in two separate variables. Use the = ,and operator to ensure that both the entered username and password must match the stored values for a successful login. Handle incorrect input using the else statement.

**`Example Input and Output:`**

 * Example 1: Successful Login

`Input:`

 * Enter username: admin

 * Enter password: 1234

**`Output:`** Login successful.

 * Example 2: Invalid Login

`Input:`

 * Enter username: user

 * Enter password: 1234

**`Output:`** Invalid credentials.

```python
# solution 5

#stored credential
stored_username = "name"

stored_password = "password"

#user input
username = input("Enter username: ")

password = input("Enter password: ")

if username == stored_username and password == stored_password:
  print("Login successful.")

else:
  print("Invalid credentials.")

```
