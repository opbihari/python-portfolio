### **8.Problem: Create a List of Even Numbers**
Imagine you want a list of even numbers between 1 and 10. Even numbers are numbers that can be divided by 2 with no remainder (like 2, 4, 6, etc.). Your goal is to create a list that includes only these even numbers within the specified range.

In this problem, you need to:

* Initialize an empty list called even_numbers to store the even numbers.
* Use range(1, 11) to generate numbers from 1 to 10.
* Use a for loop to go through each number in this range.
* Use an if condition if i % 2 == 0 to check if a number is even.
* If the number is even, add it to the even_numbers list.
* Print the resulting list of even numbers.

Key Points:
* Even numbers are numbers that can be divided by 2 without leaving a remainder (e.g., 2, 4, 6, 8).
* Start by initializing an empty list, and use a for loop and if condition to filter and add only even numbers to the list.

**Sample Output:**
Even Numbers from 1 to 10: [2, 4, 6, 8, 10]


```python
even_numbers = []
for i in range(1,11):
  if i % 2 == 0:
    even_numbers.append(i)
print(f"Even Numbers from 1 to 10 : {even_numbers}")
```

### **12. Problem: Find Average of Tuple Elements**

Imagine you have a tuple containing the prices of several items. You want to find the average price of these items to get an idea of the typical cost. The average price is calculated by adding all the prices together and then dividing by the number of items.

In this problem, you need to:

* Start with  creating a tuple of item prices.
* Use sum() to find the total of all prices.
* Divide the total by the number of prices (len() of the tuple) to get the average.
* Print the average price.

Key Points:
* The sum of the tuple’s elements gives the total of all prices.
* The len function finds the number of items in the tuple, which is used to calculate the average.

**Sample Output:**
Average Price: 250.0

```python
prices = (150,350,300,200)
total_price = sum(prices)
average_price = total_price / len(prices)

print("Average price : ",average_price)
```

### **15. Problem: Merge Two Lists of Friends**

Imagine you have two separate lists of friends. One list includes a few friends you’ve known for a while, and the other list has some new friends you recently met. You want to combine these two lists into one so that you have all your friends in a single list.

In this problem, you need to:

* Start with two lists of friends.
* Combine (or merge) these lists into one list that includes everyone from both lists.
* Print the final list to see all the friends together.

Key Points:
* The + operator allows you to join two lists into a single list.
* The order of the lists will be preserved, so the friends in the first list will appear before the friends in the second list.

**Sample Output:**
All Friends: ['Tom', 'Jerry', 'Mickey', 'Minnie']

```python
old_friends = ['Tom','Jerry']
new_friend = ['Mickey','Minnie']

all_friends = old_friends + new_friend

print(f"All Friends : {all_friends}")
```
